### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
# ferramenta de machine learning
# pip install pandas scikit-learn
# cienttista = 1
# marceneiro = 2
# pedreiro = 3

import pandas as pd


tabela = pd.read_csv("clientes.csv")
display(tabela)
display(tabela.info())

# importar só a ferramenta que precisa do sklearn
from sklearn.preprocessing import LabelEncoder

# Tranformar os dados em id:
codificador_profissao = LabelEncoder()
# fit_transform = cria os rótulos para cada coluna e substitui
tabela["profissao"] = codificador_profissao.fit_transform("profissao")

# mix_credito
codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"])

# comportamento_pagamento
codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])
```

```python

# y é quem eu vou prever
y = tabela["score_credito"]

# x é quem eu vou usar para fazer a previsão
x = tabela.drop(columns=["score_credito","id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

```

```python
# Importar IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar a IA (modelo):
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# treinar a IA (modelo)
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino,y_treino)
```

```python
# previsao
previsao_arvore = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# comparar as previsao
from  sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# Fazer nova previsão

novos_clientes = pd.read_csv("novos_clientes.csv")

display(novos_clientes)


# fit_ = diferente do fit_transform não cria os rótulos para cada coluna mas sim só substitui

# Tranformar os dados em id:
novos_clientes["profissao"] = codificador_profissao.transform("profissao")

# mix_credito
novos_clientes["mix_credito"] = codificador_mix.transform(tabela["mix_credito"])

# comportamento_pagamento
novos_clientes["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])

previsao = modelo_arvoredecisao.predict(novos_clientes)

display(novos_clientes)
display(previsao)
```
